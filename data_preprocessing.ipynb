## **Data Augmentation**

- Drop redundant columns（``City``, ``Zip code``）,since ``Latitude``, ``Logitude`` columns already represent customer location

```python
all_data = pd.read_csv('/content/drive/MyDrive/MLFP/Data/MLFP_XY(dropped).csv', index_col=0)
display(all_data)
all_data.drop(['City', 'Zip Code'], axis = 1 ,inplace=True)

display(all_data.head(5))
print(all_data.shape)
```

- Location-related columns include `` Latitude ``, ``Longitude``, ``Lat Long``, in which some values are missing but can be imputed with each other
- Impute values of``Lat Long`` to `` Latitude ``與``Longitude``, ensuring consistency of the columns' values
- After the imputation, drop column ``Lat Long``, since we have seperated lat, long columns and it is better to has the lat and long considered in 2 dimensions

```python
for i in range(all_data.shape[0]):
  if type(all_data.loc[i, 'Lat Long']) is str:  # not NAN
    s = all_data.loc[i, 'Lat Long'].index(',')
    lat = all_data.loc[i, 'Lat Long'][:s]
    longg = all_data.loc[i, 'Lat Long'][s+1:]

    all_data.loc[i, 'Latitude'] = lat
    all_data.loc[i, 'Longitude'] = longg
```

```python
all_data.drop(labels="Lat Long", axis=1, inplace=True)
all_data.Latitude = all_data.Latitude.astype(float)
all_data.Longitude = all_data.Longitude.astype(float)
display(all_data.head(5))
```

## **Encoding (non-numerical)**
- For non-numerical columns, it is necessary to encode them before training
- **Frequency encoding** is a suitable method to encode since it could represent the occurence of each categories and take rare cases into numeric considerations
- Their are two non-numerical columns that **SHOULD NOT** be frenquecy-encoded: ``Customer ID`` and ``Churn Category``
1. ``Customer ID`` is the unique identifier of each customer and does not represent any characteristic of a customer, therefore, we do not need this columns' value to predict the final result
2. ``Churn Category`` is the dependent variable, which is the variable that we want to predict, it should be simply lable encoded.

```python
numerical = set(all_data._get_numeric_data().columns)
print(numerical)
cols = all_data.columns
non_numerical = set(cols) - set(all_data._get_numeric_data().columns)
print(non_numerical)
```

```python
non_numerical.remove('Customer ID')  # no need
non_numerical.remove('Churn Category') # watch out the label
print(non_numerical)
```

### **Frequency Encoding, Lable Encoding**
1. For columns that has <= 3 unique values, that is, having under 3 unique categoris, simply use **lable encoding** to transform to 0, 1, 2
2. For columns that has > 3 categories, encode with **relative frequency**. This prevents ordinal relationship created by label encoding, which misleads the model to view the categories wrongly, for example: planA, planB, planC encoded as 0, 1, 2, suggesting planC > planB > planA. Also, label encoing does not work good on high-cardinality features and might lead to overfitting. However, frequency encoding represnet the relative frequency of each categories, which works well with tree based models.

Both encoding methods reduce the high-dimensionality curse of using the common one-hot encoding, which creates lots of features and is memory-inefficient.

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
#le.fit(["No", "Yes"])
#print(le.classes_)


for col in non_numerical:
  series = all_data[col]

  if len(series.unique()) <= 3: #label encodign
    all_data[col] = pd.Series(
        le.fit_transform(series[series.notnull()]),
        index=series[series.notnull()].index
    )
  else:
    transform = all_data[col].value_counts().to_dict() #frequency encoding
    print(transform)
    a = sum(list(transform.values()))
    for i in range(7043):
      try:
        if math.isnan(all_data.loc[i, col]):
          continue
        else:
          all_data.loc[i, col] = transform[all_data.loc[i, col]] / a #calculate relative frequency

      except:
          all_data.loc[i, col] = transform[all_data.loc[i, col]] / a
```

## **Standardization**
Standardizing the data prevents model being dominated with feature that has larger values. Though during frequency encoding, the features are encoded in relative frequencie, which are in the same scale, however, for columns that are originally numeric, it will be standardized here.

### **Min-Max Normalization**
Min-max scaling preserves the original data distribution while ensuring comparibility accross feature, unlike normal z-score standardization.

Though typically not necessary for tree-based models, which will be used later, it is not harmful to standardize.

```python
print(all_data.dtypes)
display(all_data.head(5))
```

```python
all_data["Offer"] = all_data["Offer"].astype(float)
all_data["Internet Type"] = all_data["Internet Type"].astype(float)
all_data["Contract"] = all_data["Contract"].astype(float)
all_data["Payment Method"] = all_data["Payment Method"].astype(float)
```

```python
numerical = all_data._get_numeric_data().columns
print(numerical)
print(len(numerical))

# min max standardization
for col in numerical:
    all_data[col] = (all_data[col] - all_data[col].min()) / (all_data[col].max() - all_data[col].min())
all_data.describe()
```

## **KNN imputer: fill missing value**
KNN is a good way to impute missing values, since:
1. It considered the relationship between similar instances, making this imputing method context-aware
2. Keep the original distribution of the data, while simple methods (such as: mean, mode, median imputaion) might shift the distribution
3. Can be simply applied on numeric and non-numeric data (for non-numeric data, impute with mode of neighbors)

```python
knn_input = all_data.iloc[:,1:-1]
display(knn_input.head(5))
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
new_data = imputer.fit_transform(knn_input.values)
```

```python
new_data = pd.DataFrame(new_data, columns = all_data.columns[1:-1])
display(new_data)
```

```python
new_data.insert(0, "Customer ID", all_data['Customer ID'].values)
new_data.insert(new_data.shape[1], "Churn Category", all_data['Churn Category'].values)
display(new_data.head())
```
